## 1. Importação das bibliotecas

```python
import pandas as pd
import numpy as np

# Configurações de exibição
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)

print(f'Pandas version: {pd.__version__}')
print(f'NumPy version:  {np.__version__}')
```

## 2. Carregamento do dataset

O CSV bruto fica em `data/raw/` (não versionado no Git).  
Baixar o arquivo `netflix_titles.csv` do [Kaggle](https://www.kaggle.com/datasets/shivamb/netflix-shows) e colocar nessa pasta.

```python
# Caminho do arquivo bruto (relativo à raiz do projeto)
CAMINHO_CSV = '../data/raw/netflix_titles.csv'

df = pd.read_csv(CAMINHO_CSV)
df.head()
```

## 3. Shape do dataset

Quantas linhas e colunas temos?

```python
linhas, colunas = df.shape
print(f'Total de registros: {linhas}')
print(f'Total de colunas:   {colunas}')
```

## 4. Tipos das colunas (`dtypes`)

Verificar os tipos atuais de cada coluna para identificar o que precisará ser convertido na etapa de limpeza (ex.: `date_added` está como `object` e deverá virar `datetime`).

```python
df.dtypes
```

```python
# Visão geral consolidada
df.info()
```

## 5. Valores nulos

Identificar quais colunas têm valores ausentes e em que proporção. Esse diagnóstico orienta a próxima etapa (preenchimento dos nulos).

```python
nulos = df.isnull().sum()
percentual = (df.isnull().sum() / len(df) * 100).round(2)

resumo_nulos = pd.DataFrame({
    'qtd_nulos': nulos,
    'percentual_%': percentual
}).sort_values('qtd_nulos', ascending=False)

resumo_nulos
```

## 6. Registros duplicados

Verificar a existência de duplicatas — serão removidas na etapa de limpeza.

```python
qtd_duplicados = df.duplicated().sum()
print(f'Registros duplicados: {qtd_duplicados}')
```

## 8. Inspeção rápida de colunas-chave

Olhar valores únicos de algumas colunas estratégicas para entender o que vamos tratar depois.

```python
# Tipos de conteúdo (filme vs série)
df['type'].value_counts()
```

```python
# Classificações etárias (rating)
df['rating'].value_counts(dropna=False)
```

```python
# Amostra da coluna duration — observe que mistura 'min' (filmes) e 'Seasons' (séries)
df['duration'].value_counts(dropna=False).head(15)
```

```python
# Faixa de anos de lançamento
print(f'Ano mais antigo:  {df["release_year"].min()}')
print(f'Ano mais recente: {df["release_year"].max()}')
```
